```python
#https://www.kaggle.com/code/ashydv/sales-prediction-simple-linear-regression
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
# Import the numpy and pandas package
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import matplotlib.pyplot as plt
#import pandas_datareader.data as web
import seaborn as sns
import pandas as pd
import os.path
import numpy as np
from scipy.stats import spearmanr
import time
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
start_time = time.time()

df = pd.DataFrame(pd.read_csv("titanic.csv"))
df.head()
```

```python
print(df.isnull().sum())
```

```python
#Clean
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Survived'].fillna(df['Survived'].mode()[0], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop("Cabin", axis=1,  inplace=True)
df.drop("PassengerId", axis=1,  inplace=True)
df.drop("Name", axis=1,  inplace=True)
df.drop("Ticket", axis=1,  inplace=True)


print(df.isnull().sum())
df.head()
```

```python
label_encoder = LabelEncoder()
df['Sex_encoded'] = label_encoder.fit_transform(df['Sex'])
df['Embarked_encoded'] = label_encoder.fit_transform(df['Embarked'])

# Drop the original categorical columns (we will keep the encoded versions)
df.drop(columns=['Sex', 'Embarked'], inplace=True)

# Separate features and target variable
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
print("\nLet's build our Linear regression Model\n")

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.20, random_state=1)

print("\n(X_train.head())\n")
print("======================")
print(X_train.head())

solver = 'liblinear'
lrclf = LogisticRegression(solver=solver)
lrclf = lrclf.fit(X_train, Y_train)
```

```python
print("Model Evaluation")

y_predicted = lrclf.predict(X_test)
lr_score = accuracy_score(y_predicted, Y_test)
print("Logistic Regression ({}) score: {}".format(solver, lr_score))

accuracy = accuracy_score(Y_test, y_predicted)
classification_rep = classification_report(Y_test, y_predicted)

print(f"Accuracy: {accuracy:.4f}")
print("Classification Report:")
print(classification_rep)

end_time = time.time()
time_taken = end_time - start_time
time_taken_ms = (time_taken) * 1000
print(f"Time taken for the pipeline execution: {time_taken_ms:.2f} milliseconds, {time_taken:.4f} seconds")
```
